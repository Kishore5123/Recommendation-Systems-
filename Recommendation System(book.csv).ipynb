```python
user_data = newdata.pivot_table(index='User_ID',columns='Book_title',values='Book_rating')
```

```python
user_data
```

```python
user_data.values
```

```python
user_data.fillna(0,inplace=True)
```

```python
user_data
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_data.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_data = pd.DataFrame(user_sim)
```

```python
user_sim_data.shape
```

```python
user_sim_data.index = user_data.index
user_sim_data.columns = user_data.index
```

```python
user_sim_data
```

```python
np.fill_diagonal(user_sim,0)
user_sim_data.iloc[0:5,0:5]
```

```python
user_sim_data.sort_values([8],ascending=False).head(100)
```

```python
def get_reco(User_ID):
    tem = list(user_sim_data.sort_values([User_ID],ascending=False).head(100).index)
    book_list = list(newdata[newdata["User_ID"].isin(tem)]["Book_title"])
    #print(len(book_list)
    return set(book_list)-set(newdata[newdata["User_ID"]==User_ID]["Book_title"])
    
```

```python
get_reco(10)
```
